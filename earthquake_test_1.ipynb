```python
import pandas as pd  

# file path for data
csv_file = r'C:\Users\abdif\Downloads\hot2\all_month.csv'

df = pd.read_csv(csv_file)

#print(df)

```

```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
# Depth against time according to all the datapoints

#Changed the timeformat to match the data
df['time'] = pd.to_datetime(df['time'], format='%Y-%m-%dT%H:%M:%S.%fZ')

#Plotted the graph
df.plot(x='time', y='depth', kind='line')

a = int(len(df)/5)

#reduced the number of x-axis ticks
plt.xticks(df['time'][::a])
plt.xlabel('Date')
plt.ylabel('Depth')
plt.title('Depth against Time')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D



# Create a 3D plot
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot
ax.scatter(df['dmin'], df['depth'], df['mag'], c='blue', marker='o')

# Set labels for each axis
ax.set_xlabel('Reliability of Depth (dmin)')
ax.set_ylabel('Depth')
ax.set_zlabel('Magnitude (mag)')

# Set the title of the plot
ax.set_title('3D Plot of dmin, Depth, and Magnitude')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt



# Select numerical columns from the DataFrame
numerical_columns = ['depth', 'mag', 'nst', 'gap', 'dmin', 'rms', 'horizontalError', 'depthError', 'magError', 'magNst']

# Create separate subplots for each numerical variable
fig, axes = plt.subplots(nrows=len(numerical_columns), ncols=1, figsize=(8, 2* len(numerical_columns)))

# Plot each variable in a separate subplot
for i, column in enumerate(numerical_columns):
    df[column].plot(kind='box', ax=axes[i], vert=False, patch_artist=True, medianprops={'color': 'black'})
    axes[i].set_title(f'Box Plot of {column}')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# DataFrame of type column thats counted
earthquake_types = df['type'].value_counts()

# Plotting a pie chart
plt.figure(figsize=(8, 8))
plt.pie(earthquake_types, labels=None, autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightblue', 'lightgreen', 'orange', 'black', 'yellow'])

# Adding a legend
plt.legend(earthquake_types.index, title='Earthquake Types', loc='center left', bbox_to_anchor=(1, 0.5))

plt.title('Distribution of Earthquake Types')
plt.show()
```

```python

import pandas as pd
import matplotlib.pyplot as plt

magnitude_counts = df['mag'].value_counts().sort_index()

# Plotting a bar chart
plt.figure(figsize=(10, 6))
magnitude_counts.plot(kind='bar', color='skyblue')

# Set x-axis ticks to show only 20 ticks
plt.xticks(range(0, len(magnitude_counts.index), len(magnitude_counts.index)//20), magnitude_counts.index[::len(magnitude_counts.index)//20])

plt.title('Distribution of Earthquake Magnitudes')
plt.xlabel('Magnitude')
plt.ylabel('Count')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
